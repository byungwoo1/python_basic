1차원 배열로 시리즈를 생성하고 클래스를 확인합니다.

### 시리즈 인덱싱 및 슬라이싱: iloc 인덱서

iloc(integer location) 인덱서는 정수 인덱스로 원소를 선택합니다.|

```python
sr
```

```python
# 대괄호 안에 정수 인덱스를 스칼라로 지정하면 해당 원소를 반환합니다.
sr.iloc[0]
```

```python
# 대괄호 안에 정수 스칼라 대신 리스트로 지정하면 해당 원소를 시리즈로 반환합니다.
sr.iloc[[0]]
```

```python
# 연속하지 않는 인덱스를 함께 선택하려면 반드시 리스트로 지정해야 합니다.
# 왼쪽 코드를 실행하면 sr의 0, 2번 인덱스 원소를 시리즈로 반환합니다.
sr.iloc[[0, 2]]
```

```python
# 정수 인덱스로 슬라이싱하면 연속된 원소를 시리즈로 반환합니다.
# [참고] iloc 인덱서는 끝(콜론 오른쪽) 인덱스를 포함하지 않습니다.
sr.iloc[0:2]
```

### 시리즈 인덱싱 및 슬라이싱: loc 인덱서

loc(location) 인덱서는 행이름으로 원소를 선택합니다.

```python
sr
```

```python
# 대괄호 안에 행이름을 스칼라로 지정하면 해당 원소를 반환합니다.
sr.loc['a']
```

```python
# 대괄호 안에 행이름을 스칼라 대신 리스트로 지정하면 해당 원소를 시리즈로 반환합니다.
sr.loc[['a']]
```

```python
# 연속하지 않는 행이름을 함께 선택하려면 반드시 리스트로 지정해야 합니다.
sr.loc[['a', 'c']]
```

```python
# 행이름으로 슬라이싱하면 연속된 원소를 시리즈로 반환합니다.
# [참고] loc 인덱서는 끝(콜론 오른쪽) 행이름을 포함합니다!
sr.loc['a':'c']
```

### [참고] 인덱서를 반드시 사용해야 하나?

* 인덱스가 정수 1부터 시작하는 시리즈를 생성하고 인덱싱합니다.

```python
# 인덱스가 정수 1부터 시작하는 시리즈를 생성합니다.
sr2 = pd.Series(data = range(3), index = range(1, 4))
sr2
```

sr2의 행이름에 정수 0이 없으므로 에러를 반환합니다.

sr2[0]

```python
# loc 인덱서로 sr2에서 행이름이 정수 1인 원소를 선택합니다.
sr2.loc[1]
```

* 중복 인덱스를 갖는 시리즈를 생성하고 인덱싱합니다.

```python
# 중복 인덱스를 갖는 시리즈를 생성합니다.
sr3 = pd.Series(data = range(3), index = np.tile(1, 3))
sr3
```

```python
# loc 인덱서로 sr3에서 행이름이 정수 1인 원소를 선택합니다.
sr3.loc[1]
```

```python
# iloc 인덱서로 sr3에서 0번 인덱스(첫 번째) 원소를 선택합니다.
sr3.iloc[0]
```

### 시리즈의 불리언 인덱싱

* 시리즈로 비교 연산을 실행합니다.

```python
sr
```

```python
# 시리즈로 비교 연산을 실행합니다.
sr >= 3
```

* 조건을 만족하는 원소를 선택하는 방식을 불리언 인덱싱(boolean Indexing)이라고 합니다.

```python
# 대괄호 안에 비교 연산 코드를 지정하면 True에 해당하는 원소를 선택합니다.
# [주의] iloc 인덱서를 사용하면 에러를 반환합니다.
sr.loc[sr >= 3]
```

* 두 개 이상의 조건을 고려하려면 비트 연산자를 추가합니다.

[주의] 비트 연산자의 우선순위가 비교 연산자보다 높기 때문에 아래 코드를 실행하면 정수 3과 sr의 원소 간 논리곱 연산을 먼저 실행합니다.

sr.loc[sr >= 3 & sr < 5]

```python
# [주의] 비트 연산자 양 옆 코드를 반드시 소괄호로 감싸야 합니다.
sr.loc[(sr >= 3) & (sr < 5)]
```

### 비트 연산자 사용법

부울형 시리즈 사이에 논리 연산자를 사용하면 에러를 반환합니다.

sr >= 3 and sr < 5

```python
# 논리 연산자는 진리값 사이에서 정상적으로 동작합니다.
# 따라서 numpy, pandas에서는 사용할 수 없습니다.
sr.iloc[0] >= 3 and sr.iloc[0] < 5
```

비트 연산자로 논리곱, 논리합, 배타적 논리합 또는 논리부정 연산을 실행합니다.

```python
# [논리곱] 대응하는 원소가 모두 True면 True, 아니면 False를 반환합니다.
(sr >= 3) & (sr < 5)
```

```python
# [논리합] 대응하는 원소 중 하나라도 True면 True, 아니면 False를 반환합니다.
(sr >= 3) | (sr < 5)
```

```python
# [배타적 논리합] 대응하는 원소가 다르면 True, 같으면 False를 반환합니다.
(sr >= 3) ^ (sr < 5)
```

```python
# [논리부정] True를 False로, False를 True로 반전합니다.
~ (sr >= 3)
```

### 얕은 복사 vs 깊은 복사

앝은 복사한 시리즈 원소를 변경하면 기존 시리즈 원소에 반영합니다.

```python
sr4
```

```python
# sr1을 얕은 복사한 sr4를 생성합니다.
sr4 = sr1
sr4
```

```python
# sr4의 첫 번째 원소를 2로 변경합니다.
sr4.iloc[0] = 2
sr4
```

```python
# sr1의 첫 번째 원소도 바뀌었습니다.
# [참고] 두 변수가 같은 객체를 가리키고 있기 때문입니다.
sr1
```

깊은 복사한 시리즈는 기존 시리즈와 별개이므로 변경 사항을 반영하지 않습니다.

```python
# sr1을 깊은 복사한 sr5를 생성합니다.
sr5 = sr1.copy()
sr5
```

```python
# sr5의 첫 번째 원소를 1로 변경합니다.
sr5.iloc[0] = 1
sr5
```

```python
# sr1의 첫 번째 원소는 바뀌지 않았습니다.
# [참고] 두 변수가 가리키는 객체가 다르기 때문입니다.
sr1
```

### 데이터프레임 생성

데이터프레임은 2차원 배열, 리스트 또는 딕셔너리로 생성하며 인덱스(행이름)와 컬럼명(열이름)을 갖습니다.

```python
# 대괄호 안에 행이름을 스칼라로 지정하면 해당 행을 시리즈로 반환합니다.
df.loc[1]
```
