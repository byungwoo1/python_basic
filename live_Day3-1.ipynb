### 데이터프레임 병합

```python
set(price.columns) & set(detail.columns) # 교집합이 있는지 확인. 
```

```python
pd.merge(left=price,
        right=detail,
        how='inner',# 병합 방법 : inner 내부 병합
        left_on='주소',
        right_on='지번주소')
```

```python
detail=detail.rename(columns={'지번주소':'주소'})
```

```python
set(price.columns) & set(detail.columns) #하기 일치하는 이름이 있는 경우
```

```python
apt=pd.merge(left=price,
             right=detail,
             how='inner',
             on='주소')
```
