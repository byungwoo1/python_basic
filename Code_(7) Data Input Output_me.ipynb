## 새 텍스트 파일 생성

## 쓰기모드로 텍스트 파일을 엽니다.

```python
file = open(file = 'test.txt', mode = 'w', encoding = 'UTF-8')
# 현재 작업 경로에 있는 'test.txt' 파일을 쓰기 모드로 엽니다. 파일이 없으면 새로 생성합니다.
# [참고] Windows 사용자는 encoding을 추가하지 않으면 'CP949'를 자동 적용합니다
```

## 텍스트 파일에 문자열을 입력합니다.

```python
for i in range(1, 6): # 반복문을 실행하여 여러 문자열을 텍스트 파일에 입력합니다.
    file.write(f'{i} 페이지 수집!\n')
```

```python
# 텍스트 파일을 닫습니다.
file.close() # 텍스트 파일을 닫으면 텍스트 파일을 저장합니다.
```

## 기존 텍스트 파일에 문자열 추가

```python
# 추가모드로 텍스트 파일을 엽니다.
file = open(file = 'test.txt', mode = 'a', encoding = 'UTF-8')
# 현재 작업 경로에 있는 'test.txt' 파일을 추가모드로 엽니다. 파일이 없으면 새로 생성합니다.
```

```python
# 텍스트 파일에 문자열을 입력합니다.
for i in range(11, 16):
    file.write(f'{i} 페이지 수집!\n')
```

```python
# 텍스트 파일을 닫습니다.
file.close()
```

## 텍스트 파일을 문자열로 읽기

```python
# 문자열 읽기모드로 텍스트 파일을 엽니다.
file = open(file = 'test.txt', mode = 'r', encoding = 'UTF-8')
# 현재 작업 경로에 있는 'test.txt' 파일을 읽기모드로 엽니다. 파일이 없으면 에러를 반환합니다.
```

```python
# file을 읽고 text에 할당합니다.
text = file.read() # read() 함수는 file을 사람이 읽을 수 있는 문자열(str)로 반환합니다.
                   # [참고] readlines() 함수는 리스트로 반환합니다.
text
```

```python
# text의 클래스를 확인합니다. text의 클래스는 str입니다.
# [참고] str은 사람이 읽을 수 있는 문자열입니다.
type(text)
```

```python
# 문자열을 인코딩하면 바이너리 코드로 변환합니다.
text.encode(encoding = 'UTF-8')
```

## 텍스트 파일을 바이너리로 읽기

### 바이너리 읽기모드로 텍스트 파일을 엽니다.

```python
# [주의] 바이너리 모드로 읽을 때 인코딩 방식을 추가할 수 없습니다.
file = open(file = 'test.txt', mode = 'rb')
# 현재 작업 경로에 있는 'test.txt' 파일을 바이너리 읽기모드로 엽니다. 파일이 없으면 에러를 반환합니다.
```

### 텍스트 파일을 읽고 출력합니다.

```python
# file을 읽고 text에 할당합니다.
text = file.read()
text
```

```python
# text의 클래스를 확인합니다. text의 클래스는 bytes입니다.
# [참고] bytes는 사람이 읽을 수 없는 코드 입니다.
type(text)
```

```python
# 바이너리 코드를 디코딩하면 문자열로 변환합니다.
text.decode(encoding = 'UTF-8')
# [주의] Windows에서 'test.txt' 파일을 생성할 때 encoding = 'UTF-8'를 추가하지 않았다면
# 반드시 'EUC-KR' 또는 'CP949'를 지정해야 합니다.
```

### csv 파일명을 재사용할 수 있도록 변수에 할당합니다.

```python
file = 'KBO_Hitters_2021.csv'
```

### csv 파일을 읽고 데이터프레임을 생성합니다.

###df = pd.read_csv(filepath_or_buffer = file)
### [주의] csv 파일의 인토딩 방식이 'UTF-8'이 아니면 인코딩 방식을 지정해야 합니다.

### csv 파일을 바이너리 모드로 읽습니다.

```python
text = open(file = file, mode = 'rb').read()
# [참고] text는 bytes 코드이므로 출력하면 사람이 읽을 수 없습니다.
```

### csv 파일의 문자 인코딩 방식을 확인합니다.

```python
chardet.detect(text)
# [참고] 바이너리 코드의 인토딩 방식과 신뢰도(confidence)를 반환합니다.
```

### csv 파일의 인코딩 방식을 지정하여 다시 읽습니다.

```python
df = pd.read_csv(filepath_or_buffer = file, encoding = 'EUC-KR')
```

```python
df.index = range(1, df.shape[0] + 1) #[참고] df.shape은 행과 열 개수를 튜플로 반환합니다.
```

```python
# df의 처음 5행을 출력합니다.
df.head()
```

### 데이터프레임을 csv 파일로 저장합니다.

```python
df.to_csv(path_or_buf = 'test.csv', index = None, encoding = 'UTF-8')
# [참고] csv 파일의 인코딩 방식을 지정할 수 있습니다.
```

### [참고] 실습 파일 삭제

### 현재 작업 경로에서 'test'를 포함하는 파일명으로 리스트를 생성합니다.

```python
files = [file for file in os.listdir() if 'test' in file]
files # files는 이번 시간에 생성한 4개 파일명을 원소로 갖습니다.
```

### 반복문으로 해당 파일을 모두 삭제합니다.

```python
for file in files:
    os.remove(path = file) # [참고] os.remove() 함수는 한 번에 하나의 파일을 삭제합니다.
```
